```python
import pandas as pd
```

```python
import glob

# Step 1: Define the folder path where your CSVs are located
folder_path = "/kaggle/input/sales-data"  # Replace with your dataset folder

# Step 2: Get all CSV files in the folder
csv_files = glob.glob(f"{folder_path}/*.csv")

# Step 3: Read and merge all CSV files
df = pd.concat((pd.read_csv(file) for file in csv_files), ignore_index=True)

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
# Remove duplicates while keeping the first occurrence
df = df.drop_duplicates(keep='first')
```
